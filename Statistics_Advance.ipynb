```python
# Ques-9  Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample
#size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?
#Task: Write Python code to calculate the z-score and p-value for the given data.

# Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.



```

```python
import numpy as np
import scipy.stats as stats

# Given data
sample_mean = 105
population_mean = 100
population_std_dev = 15
sample_size = 25
alpha = 0.05

# Calculate the z-score
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate the p-value for a one-tailed test
p_value = 1 - stats.norm.cdf(z_score)

# Print the results
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision on the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

```

```python
# Ques 10- Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.
# Generate 1,000 samples and plot the distribution. What is the expected mean and variance?
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10  # number of trials
p = 0.6  # probability of success
num_samples = 1000  # number of samples

# Generate binomial samples
samples = np.random.binomial(n, p, num_samples)

# Calculate expected mean and variance
expected_mean = n * p
expected_variance = n * p * (1 - p)

# Plot the distribution
plt.figure(figsize=(10, 5))
sns.histplot(samples, bins=range(0, n + 2), stat='density', kde=False, color='blue')
plt.title('Simulated Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Density')
plt.xticks(range(0, n + 1))
plt.grid(axis='y')
plt.axvline(expected_mean, color='red', linestyle='--', label='Expected Mean')
plt.legend()
plt.show()

# Print expected mean and variance
print(f'Expected Mean: {expected_mean:.2f}')
print(f'Expected Variance: {expected_variance:.2f}')

```
